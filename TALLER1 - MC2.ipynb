```python
def NUDFT(y,t,f,N):
    d =0
    for i in range(N):
        c=y[i]*np.exp(-2*np.pi*1j*t[i]*f)
        d += c
        
    return d
```

```python
def senal(t):
    senal=np.sin(2*np.pi*np.sqrt(2)*t)
    
    return senal
```

```python
def NUDFT(y,t,f,N):
    d =0
    for i in range(N):
        c=y[i]*np.exp(-2*np.pi*1j*t[i]*f)
        d += c
    d = abs(d) 
    
    return d
```

```python
t= np.linspace(0,20,200)
N = len(t)
dt = 20/200
y= senal(t)
f = np.linspace(0, 1/(dt),200)

TFsenal=NUDFT(y,t,f,N)
print(TFsenal)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(t,TFsenal, color ='blue')
plt.xlabel('Frecuencia(Hz)')
plt.ylabel('TF')
plt.show()
```

```python
def senal_ruido(t):
    senal=np.sin(2*np.pi*np.sqrt(2)*t+np.random.normal(0,0.1,N))
    
    return senal
```

```python
t= np.linspace(0,20,200)
N = len(t)
dt = 20/200
yr= senal_ruido(t)
f = np.linspace(0, 1/(dt),200)

TFsenal=NUDFT(yr,t,f,N)
print(TFsenal)
```

```python
plt.plot(t,TFsenal, color ='#7BC8F6')
plt.xlabel('Frecuencia(Hz)')
plt.ylabel('TF')
plt.show()
```

```python
def NUDFT(y,t,f,N):
    d =0
    for i in range(N):
        c=y[i]*np.exp(-2*np.pi*1j*t[i]*f)
        d += c
    d = abs(d) 
    
    return d
```

```python
t= np.linspace(0,20,200)
N = len(t)
dt = 20/200
y= senal(t)
f = np.linspace(0, 1/(dt),200)

TFsenal=NUDFT(y,t,f,N)
print(TFsenal)
```

```python
y_max = TFsenal.max()
print (y_max/2)
```

```python
FWHM = y_max/2
t= np.linspace(0,20,200)
N = len(t)
dt = 20/200
y= senal(t)
f = np.linspace(0, 1/(dt),200)

TFsenal=NUDFT(y,t,f,N)
print(TFsenal)
```

```python
TFsenal_ = TFsenal[1:100]
idx = abs(TFsenal_ - (y_max/2))
idx = idx.argmin()
print(idx)
```

```python
FWHM = (f[idx]-f[TFsenal_.argmax()])
print(FWHM)
```

```python
FWHM = y_max/2
N= 500
t_max= 1000
t= np.linspace(0,t_max,N)
dt = t_max/N
y= senal(t)
f = np.linspace(0, 1/(dt),N)

TFsenal=NUDFT(y,t,f,N)
print(TFsenal)
```

```python
plt.plot(f,TFsenal, color ='#7BC8F6')
plt.xlabel('Frecuencia(Hz)')
plt.ylabel('TF')
plt.show()
```

Al aumentar el nmero de muestras los picos se encuentran en frecuencias ms bajas y los picos son ms finos, al disminuir el npmero de muestras los picos se encuentran en frecuencias ms grandes y los picos son ms anchos. Al aumentar el tiempo mximo los picos se hacen ms cercanos en la frecuencia y al disminuirlo, se alejan.
